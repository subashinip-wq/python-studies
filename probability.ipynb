```python
import numpy as np

# Prior probabilities
prior = np.array([0.50, 0.30, 0.20])

# Likelihood probabilities
likelihood = np.array([0.40, 0.70, 0.20])

# Total probability
P_L = np.sum(prior * likelihood)

# Posterior probabilities using Bayes' theorem
posterior = (prior * likelihood) / P_L

# Display posterior probabilities
print("Posterior Probabilities:", posterior)

# Causes
causes = ["Congestion", "Hardware Failure", "Signal Interference"]

# Most likely cause
print("Most Likely Cause:", causes[np.argmax(posterior)])
```
